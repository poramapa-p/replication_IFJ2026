```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## import all functions from DensityEst 
## see DensityEst.ipynb for detailed explanations 

from DensityEst import GeneralizedBetaEst,GeneralizedBetaStats,TriangleEst,TriangleStats,UniformEst,UniformStats,SynDensityStat
```

```python
### loading probabilistic data  
GDPSPF=pd.read_stata('C:/Users/copyl/OneDrive/MU/paper1/Replication/data/us_spf_202506.dta')
GDPSPF['YQ'] = pd.to_datetime(GDPSPF['year'].astype(str) + 'Q' + GDPSPF['quarter'].astype(str))
print(GDPSPF.YQ)
# SPF annual RGDP 
# horizon: y to y+1
```

```python
## survey-specific parameters
# upper bound of bins
bin_9209=np.array([-np.Inf, -1.1, -0.1, 0.9, 1.9, 2.9, 3.9, 4.9, 5.9, np.Inf])
index_9209 = GDPSPF.index[(GDPSPF.YQ>='1992Q1')&(GDPSPF.YQ<='2009Q1')]
bin_0920=np.array([-np.Inf, -2.1, -1.1, -0.1, 0.9, 1.9, 2.9, 3.9, 4.9, 5.9, np.Inf])
index_0920 = GDPSPF.index[(GDPSPF.YQ>='2009Q2')&(GDPSPF.YQ<='2020Q1')]
bin_2023=np.array([-np.Inf, -6.1, -3.1, -0.1, 1.4, 2.4, 3.9, 6.9, 9.9, 15.9, np.Inf])
index_2023 = GDPSPF.index[GDPSPF.YQ>='2020Q2']
```

```python
##############################################
### attention: the estimation happens here!!!!!
###################################################

## creating positions 
index  = GDPSPF.index
columns=['mean','var', 'std']
GDPSPF_moment_est = pd.DataFrame(index=index,
                                 columns=columns)

## Invoking the estimation

for i in index_9209:
    print(i)
    ## take the probabilities (flip to the right order, normalized to 0-1)
    PRGDP_y1 = np.flip(np.array([GDPSPF.iloc[i,:]['prgdp'+str(n)]/100 for n in range(11,20)]))
    print(PRGDP_y1)
    if not np.isnan(PRGDP_y1).all():
        stats_est=SynDensityStat(bin_9209, PRGDP_y1)
        if  stats_est is not None and len(stats_est):
            print(stats_est['mean'])
            GDPSPF_moment_est['mean'][i]=stats_est['mean']
#             print(stats_est['variance'])
            GDPSPF_moment_est['var'][i]=stats_est['variance']
            print(stats_est['std'])
            GDPSPF_moment_est['std'][i]=stats_est['std']
        else:
            GDPSPF_moment_est['mean'][i] = np.nan
            GDPSPF_moment_est['var'][i] = np.nan
            GDPSPF_moment_est['std'][i] = np.nan
            
for i in index_0920:
    print(i)
    ## take the probabilities (flip to the right order, normalized to 0-1)
    PRGDP_y1 = np.flip(np.array([GDPSPF.iloc[i,:]['prgdp'+str(n)]/100 for n in range(12,22)]))
    print(PRGDP_y1)
    if not np.isnan(PRGDP_y1).all():
        stats_est=SynDensityStat(bin_0920, PRGDP_y1)
        if  stats_est is not None and len(stats_est):
            print(stats_est['mean'])
            GDPSPF_moment_est['mean'][i]=stats_est['mean']
#             print(stats_est['variance'])
            GDPSPF_moment_est['var'][i]=stats_est['variance']
            print(stats_est['std'])
            GDPSPF_moment_est['std'][i]=stats_est['std']
        else:
            GDPSPF_moment_est['mean'][i] = np.nan
            GDPSPF_moment_est['var'][i] = np.nan
            GDPSPF_moment_est['std'][i] = np.nan
            

for i in index_2023:
    print(i)
    ## take the probabilities (flip to the right order, normalized to 0-1)
    PRGDP_y1 = np.flip(np.array([GDPSPF.iloc[i,:]['prgdp'+str(n)]/100 for n in range(12,22)]))
    print(PRGDP_y1)
    if not np.isnan(PRGDP_y1).all():
        stats_est=SynDensityStat(bin_2023, PRGDP_y1)
        if  stats_est is not None and len(stats_est):
            print(stats_est['mean'])
            GDPSPF_moment_est['mean'][i]=stats_est['mean']
#             print(stats_est['variance'])
            GDPSPF_moment_est['var'][i]=stats_est['variance']
            print(stats_est['std'])
            GDPSPF_moment_est['std'][i]=stats_est['std']
        else:
            GDPSPF_moment_est['mean'][i] = np.nan
            GDPSPF_moment_est['var'][i] = np.nan
            GDPSPF_moment_est['std'][i] = np.nan
```

```python
### exporting moments estimates to pkl
GDPSPF_est = pd.concat([GDPSPF,GDPSPF_moment_est], join='inner', axis=1)
GDPSPF_est.to_pickle("./GDPSPFDst.pkl")
GDPSPF_pk = pd.read_pickle('./GDPSPFDst.pkl')

## convert all to numeric 
type_map = {columns[i]:'float' for i,column in enumerate(columns)}
GDPSPF_pk =GDPSPF_pk.astype(type_map)
GDPSPF_pk.columns

## export to stata 

GDPSPF_pk.to_stata('C:/Users/copyl/OneDrive/MU/paper1/Replication/data/us_spf_est_202506.dta')
```
